# Frequentist vs Bayesian visualization

```python
# import pymc3 to use HPD calculation
import pymc3 as pm

data_model = Semisync5bees()
samples_count = 10
p_true = [0.3, 0.25, 0.35, 0.45, 0.5]
p_tscc_true = data_model.eval_bscc_pfuncs(p_true)

# Distance function
def rmse(p, p_hat):
    s = 0
    for it in zip(p, p_hat):
        diff = it[0] - it[1]
        s += diff * diff
    return math.sqrt(s / len(p))

# Frequentist interval
import math
def get_interval(d, trials_count):
    intervals = []
    for di in d:
        z_a2 = 1.96
        margin = z_a2 * math.sqrt(di* (1 - di) / trials_count) + 0.5 / trials_count
        ub = di + margin
        lb = di - margin
        intervals.append((ub, lb))
    return intervals

# Bayes interval
def bayes_step(data, alpha):
    new_alpha = [sum(x) for x in zip(alpha, data)]
    estimated_p = alpha / sum(alpha)
    return (new_alpha, estimated_p)

def bayes(data):
    bins_count = len(data[0])
    alpha = np.ones(bins_count)
    estimated_p = np.zeros(bins_count)
    traces = []
    for i in range(0, len(data)):
        alpha, estimated_p = bayes_step(data[i], alpha)
        posterior_sample = np.random.dirichlet(alpha, 1000)
        bayes_intervals = pm.stats.hpd(posterior_sample, credible_interval=0.95)
        traces.append((estimated_p, bayes_intervals))
    return traces

def frequentist(multinomial_samples):
    freq_results = []
    trials_count = 0
    bins_count = len(multinomial_samples[0])
    m = np.zeros(bins_count)
    for d in multinomial_samples:
        m = m + d 
        trials_count = sum(m)
        estimated_p = m / sum(m)
        intervals = get_interval(estimated_p, trials_count)
        freq_results.append((estimated_p, intervals))
    return freq_results

def synthesize_data(model, trials_count, samples_count):
    data = []
    p_tscc_true = model.eval_bscc_pfuncs(p_true)
    for i in range(0, samples_count):
        s = model.sample(params=p_true, trials_count=trials_count)
        data.append(s)
    return data, p_tscc_true

def estimate_params(model, samples_count):
    trials_count = 10
    data, p_tscc_true = synthesize_data(model, trials_count, samples_count)
    multinomial_samples = [d[1] for d in data]
    #Frequentist
    freq_results = frequentist(multinomial_samples)
    #Bayes
    bayes_results = bayes(multinomial_samples)
    return (freq_results, bayes_results, data)
    
def visualize_point(p_true, p_f_hat, p_b_hat):
    fig = plt.gcf()
    fig.set_size_inches(8.5, 5.5)
    plt.plot(p_true, 'rx')
    plt.plot(p_f_hat, 'g.')
    plt.plot(p_b_hat, 'b.')
    caption = "Point estimation, green is frequentist, blue is bayesian"
    plt.figtext(0.5, 0.01, caption, wrap=True, horizontalalignment='center', fontsize=12)
    plt.show()

def visualize_intervals(freq_intervals, bayes_intervals, p_f_hat, p_b_hat, p_tscc_true):
    fig = plt.gcf()
    fig.set_size_inches(8.5, 5.5)
    plt.plot(p_tscc_true, 'r.')
    plt.plot(p_f_hat, 'g_')
    plt.plot(p_b_hat, 'bx')
    for i, it in enumerate(zip(freq_intervals, bayes_intervals)):
        f = it[0]
        b = it[1]
        offset = 0.1
        plt.plot([i + offset, i + offset], [f[0], f[1]], 'g-')
        offset = 0.2
        plt.plot([i + offset, i + offset], [b[0], b[1]], 'b-')
    caption = "Interval estimation, green is frequentist, blue is bayesian"
    plt.figtext(0.5, 0.01, caption, wrap=True, horizontalalignment='center', fontsize=12)
    plt.show()
    
estimation_results = estimate_params(data_model, samples_count)

def visualize_results(result):
    f_result = result[0]
    b_result = result[1]
    data = result[2]
    i = 0
    for res in zip(f_result, b_result):
        print("##### Data batch {}#####".format(i))
        print("Histogram {}".format(data[i][1]))
        p_f_hat = res[0][0]
        p_b_hat = res[1][0]
        freq_intervals = res[0][1]
        bayes_intervals = res[1][1]
        print("Distribution (true) {}".format(p_tscc_true))
        print("Distribution (in this data batch)    {}".format(data[i][2]))
        print("Distribution (from accumulated data) {}".format(p_f_hat))
        i += 1

        #visualize_point(p_tscc_true, p_f_hat, p_b_hat)
        visualize_intervals(freq_intervals, bayes_intervals, p_f_hat, p_b_hat, p_tscc_true)
        
visualize_results(estimation_results)

#100: 10 - 10 - 10 ... - 10 
#     p1   p2  
```
