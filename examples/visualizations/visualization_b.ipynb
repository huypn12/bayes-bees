```python
from sklearn.metrics import mean_squared_error
from math import sqrt

import numpy as np

# Plotting
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl

def rmse(y_actual, y_predicted):
    rms = sqrt(mean_squared_error(y_actual, y_predicted))
    return rms
```
