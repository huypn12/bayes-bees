```python
def visualize_hpds(hpd_list, p_hat_list, p_true):
    fig = plt.gcf()
    fig.set_size_inches(16, 10)
    plt.plot(p_true, 'rx')
    plt.xlabel("Chain parameter index")
    plt.ylabel("Parameter's domain")
    plt.xticks(np.arange(len(p_true)))
    for i in range(0, len(p_true)):
        offset = 0.1
        for hpd, p_hat in zip(hpd_list, p_hat_list):
            hpd_p_i = hpd[i]
            l = hpd_p_i[0]
            u = hpd_p_i[1]
            plt.plot(i + offset, p_hat[i], 'g_')
            plt.plot([i + offset, i + offset], [l, u], 'g-')
            offset += 0.15
    for xc in np.arange(len(p_true)):
        plt.axvline(x=xc, linewidth=0.5)
    caption = ""
    plt.figtext(0.5, 0.01, caption, wrap=True, horizontalalignment='center', fontsize=12)
    plt.show()
```

```python
def visualize_hpds_2(hpd_list, p_hat_list, p_true):
    fig = plt.gcf()
    fig.set_size_inches(12, 10)
    custom_lines = [Line2D([0], [0], color='green', lw=4),
                    Line2D([0], [0], color='magenta', lw=4)]
    _, ax = plt.subplots()
    ax.legend(custom_lines, ['Rational functions', 'DTMC sampling'])
    plt.plot(p_true, 'rx')
    plt.xlabel("Chain parameter index")
    plt.ylabel("Parameter's domain")
    plt.xticks(np.arange(len(p_true)))
    j = 1
    for i in range(0, len(p_true)):
        offset = 0.2
        for hpd, p_hat in zip(hpd_list, p_hat_list):
            hpd_p_i = hpd[i]
            l = hpd_p_i[0]
            u = hpd_p_i[1]
            color = 'g' if j % 2 != 0 else 'm'
            plt.plot(i + offset, p_hat[i], color + '_')
            plt.plot([i + offset, i + offset], [l, u], color + '-')
            offset += 0.3
            j += 1
    for xc in np.arange(len(p_true)):
        plt.axvline(x=xc, linewidth=0.5)
    caption = ""
    plt.figtext(0.5, 0.01, caption, wrap=True, horizontalalignment='center', fontsize=12)
    plt.show()
```
